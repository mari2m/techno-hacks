```python
import matplotlib.pyplot as plt
import seaborn as sns
for column in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for column in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plotting for categorical columns
for column in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df[column])
    plt.title(f'Count of Categories in {column}')
    plt.ylabel('Count')
    plt.show()
```

```python
for i in ['Age', 'Fare']:
    died = list(df[df['Survived'] == 0][i].dropna())
    survived = list(df[df['Survived'] == 1][i].dropna())
    xmin =min(min(died), min(survived))
    xmax = max(max(died ), max(survived))
    width = (xmax -xmin) / 40
    sns.distplot(died, color = 'r', kde = False, bins= np.arange(xmin, xmax, width ))
    sns.distplot(survived, color = 'g', kde = False, bins = np.arange(xmin,xmax, width))
    plt.legend(['Did not survive', 'Survived'])
    plt.title('Overlaid histogram for {}'.format(i))
    plt.show()
```

```python
# Gender countplot visualization
sns.displot(x=df['Sex'], hue=df['Survived'])
plt.show()
```

```python
sns.displot(df, x="Survived", hue="Cabin", element="step")
plt.show()
```

```python
df['Cabin'] = df['Cabin'].astype(str)
df['Survived'] = df['Survived'].astype(str)
copy_df = df.copy()
copy_df['Cabin_Group'] = df['Cabin'].str[0]

plt.figure(figsize=(7,4))
sns.countplot(x=copy_df['Cabin_Group'], hue=copy_df['Survived'], palette='deep')
plt.title("# of Survivors in Each Cabin Group")
plt.tight_layout()
plt.show()
```
